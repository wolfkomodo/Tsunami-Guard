* Create a function to return a list of the first n odd numbers like so: `def get_odds(n):` Be sure to test it!
* Create a function to return a list of the first n square numbers like so: `def get_squares(n):` Be sure to test it!
* Create a list of tuples [($x_1, y_1), (x_2, y_2$) ...] of length 10 where $x_n$ represents the nth odd number and $y_n$ represents the nth square. The first two elements would be (1,1) and (3, 4).
* Create a scatterplot of the 10 points in the list from above, where (x,y) corresponds to 1 point. For instance, (1,1) and (15, 64) should both be on the plot.
* Write a function to transpose a list of lists (i.e. a rectangular matrix). Transposing is making the columns into rows and vice versa. For example, transpose([[1, 2, 3], [4, 5, 6]]) should return [[1, 4], [2, 6], [3, 6]]. Be sure to test it!

```python
# Implement TODO 5 here:

print('Focus level: %s' % project['plan']['focus_level'])

def get_odds(n):
    return [2*i+1 for i in range(n)]

def get_squares(n):
    return [i**2 for i in range(1, n+1)]

odds = get_odds(10)
squares = get_squares(10)
print(list(zip(odds, squares)))
plt.scatter(odds, squares);

# There are MUCH better ways to do this - I tried to write this in a simple way because you're just starting
def transpose(matrix):
    # Define an blank matrix to return
    new_matrix = []
    # Iterate the number of number of rows in the new_matrix (same as the number of columns in matrix)
    for i in range(len(matrix[0])):
        # Define a blank row of the new_matrix
        new_row = []
        # Iterate over each row in the matrix
        for lst in matrix:
            # Access the row number as column index (transposing!)
            new_row.append(lst[i])
        # Add the newly made row to the new_matrix
        new_matrix.append(new_row)
    return new_matrix
```
