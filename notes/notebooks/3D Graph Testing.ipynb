```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random
import numpy as np
from matplotlib.figure import Figure
```

```python
lst = [i for i in range(0, 10)]
lst[int(2.3)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
blue_num = 0
brown_num = 0
coll = ax.scatter(1, 1, 1, c='brown', marker='o')
for x in range(1, 101):
    for y in range(1, 101):
        for z in range(1, 101):
            a = random.randint(1, 60)
            b = random.randint(1, 60)
            if z > a*y + b:
                coll.append(ax.scatter(x, y, z, c='blue', marker='o'))
                blue_num += 1
            else:
                coll.append(ax.scatter(x, y, z, c='brown', marker='o'))
                brown_num += 1
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim3d(0, 100);
ax.set_ylim3d(0, 100);
ax.set_zlim3d(0, 100);
print('Number of blue dots: %d' % blue_num)
print('Number of brown dots: %d' % brown_num)
```
